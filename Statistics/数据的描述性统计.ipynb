```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## 数据读取

数据来源于mysql官网测试数据,这里抽取了salary表10w条数据：https://dev.mysql.com/doc/employee/en/

```python

salary_df = pd.read_csv('./employees_salaries_10w.csv')
salary = salary_df['salary']
salary.values.sort()

# 数据分布情况
salary.plot(figsize=(10,5))
```

```python
# 大体数据统计
salary.describe()
```

## 集中趋势

```python
print(u'算术平均数: ', salary.mean())
print(u'中位数: ', salary.median())
print(u'众数: ', salary.mode())
print(u'众数: ', salary.mode())
print(u'最小值', salary.min()) 
print(u'最大值', salary.max())
```

```python
# 分位数
print (u'第一分位数: ', salary.quantile(q=0.25))
print (u'第二分位数: ', salary.quantile(q=0.50))
print (u'第三分位数: ', salary.quantile(q=0.75))
```

## 离中趋势

```python
print(u'方差: ', salary.var())
print(u'标准差: ', salary.std())
print(u'四分位差: ', salary.quantile(q=0.75)-salary.quantile(q=0.25))
print(u'离散系数=标准差/均值: ', salary.std()/salary.mean())
```

```python
# hist
plt.hist(salary,100)
```

```python
print (u'偏态系数: ', salary.skew())
print (u'峰态系数: ', salary.kurt())
```
