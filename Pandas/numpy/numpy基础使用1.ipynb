### 布尔型索引

```python
name = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.rand(7, 4)
print('\n name: ', name)
print('\n data: ', data)

# 使用name array产生的bool值索引data对应的数据。要求布尔型数组的长度必须和被索引的轴长度一致。
print("\n name == 'Bob': ", name == 'Bob')
print("\n data[name == 'Bob']: ", data[name == 'Bob'])

# 布尔索引后，切分其他轴
print("\n data[name == 'Bob']: ", data[name == 'Bob', 2:])

# 多个布尔条件组合
print("\n data[(name == 'Bob') | (name=='Will')]: ", data[(name == 'Bob') | (name=='Will')])
# 切忌写成这样： print("\n data[name == 'Bob']", data[name == 'Bob' | name=='Will']) 会存在运算符优先级问题： numpy “TypeError: ufunc 'bitwise_and' not supported for the input types” when using a dynamically created boolean mask

# (重点)通过布尔数组设置是一种经常用到的手段
print("\n data[data>=0.5]: ", data[data>=0.5])

# 索引后设置数据
data[name!='Joe']=7
print("\n data[name!='Joe']=7: ", data)
```

### 花式索引
使用整数数组进行索引

```python
arr = np.empty((8, 4))

for i in range(8):
    arr[i] = i
print("arr: ", arr)
```

```python
# 使用整数数组进行索引
print(arr[[4,0,1,5]])
```

```python
arr = np.arange(32).reshape((8,4))
print("\n arr: ", arr)

# 精确索引
print("\n arr[[1,5,7,2], [0,3,1,2]]: ", arr[[1,5,7,2], [0,3,1,2]])

# 块索引
print("\n arr[[1,5,7,2]][:, [0,3,1,2]]): ", arr[[1,5,7,2]][:, [0,3,1,2]])
print("\n arr[np.ix_([1,5,7,2], [0,3,1,2])]): ", arr[np.ix_([1,5,7,2], [0,3,1,2])])
```

## 将条件逻辑转化为数组运算
where可以写出很多复杂的逻辑，支持嵌套

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])

result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
print("result: ", result)

# 以上非矢量化，使用循环的方式有两个缺点：1.处理速度慢(纯python运行)；2.无法用于多维数组
result = np.where(cond, xarr, yarr)
print("result: ", result)
```

```python
arr = np.random.randn(4,4)
print("\n arr: ", arr)

print("\n np.where(arr>0, 2, -2): ", np.where(arr>0, 2, -2))
```

## 排序
顶级方法np.sort放回的是数组的副本，而就地排序会修改数组本身

```python
# 一维数组排序
arr  = np.random.randn(8)
print("arr: ", arr)

arr.sort()
print("after sort: ", arr)
```

```python
# 多维数组排序
# 指定任意一个轴进行排序
arr = np.random.randn(5,3)
print("arr: ", arr)

arr.sort(1)
print("after sort: ", arr)
```

```python
# 计算数组分位数最简单的办法就是对其进行排序，然后取特定位置的值

large_arr = np.random.randn(1000)
large_arr.sort()
print(large_arr[int(0.05 * len(large_arr))])
```

## 随机数生成

```python
# 均匀分布的样本值
arr = np.random.rand(100)
df = pd.DataFrame(arr)
df.plot()
```

```python
# 给定上下限范围内随机选取整数
arr = np.random.randint(low=0, high=10000, size=500)
df = pd.DataFrame(arr)
df.plot()
```

```python
# 正态分布(平均值为0，标准差1)
arr = np.random.randn(100)
df = pd.DataFrame(arr)
df.plot()
```

```python
# 二项分布
arr = np.random.binomial(10, 0.5, 100)
df = pd.DataFrame(arr)
df.plot()
```

```python
# 正态(高斯)分布
arr = np.random.normal(3, 2.5, size=100)
df = pd.DataFrame(arr)
df.plot()
```

```python
# beta分布
arr = np.random.beta(3, 2.5, size=100)
df = pd.DataFrame(arr)
df.plot()
```

```python
# 卡方分布
arr = np.random.gamma(3, 2.5, size=100)
df = pd.DataFrame(arr)
df.plot()
```
