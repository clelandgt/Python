```python
# pandas常用模块
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from numpy.random import randn

# 使用ipython效果会更好，会打开对话窗口，显示效果更好。而jupyter上只是一个截图
```

```python
x = np.arange(1000)
y = x * 2 + np.random.randint(-200,200, size=1000)
```

```python
plt.figure(figsize=(4, 8))
# alpha: 透明度
plt.scatter(x, y, alpha=0.2)
# 设置x,y轴
plt.yticks(np.arange(0,2500,200))
plt.xticks(np.arange(0,1000,200))
plt.show()
```
